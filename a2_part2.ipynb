# Speech Denoising

Import libraries

```python
import librosa
import numpy as np
import tensorflow as tf
```

Load the audio files, apply fourier transform and take magnitude of it. 

```python
s, sr=librosa.load("train_clean_male.wav", sr=None)
S=librosa.stft(s, n_fft=1024, hop_length=512)
sn, sr=librosa.load("train_dirty_male.wav", sr=None)
X=librosa.stft(sn, n_fft=1024, hop_length=512)
S_abs=np.abs(np.transpose(S))
X_abs=np.abs(np.transpose(X))
```

Define the model with two hidden layers each containing 1024 neurons

```python
x = tf.placeholder(tf.float32, [None, 513])
h1= tf.layers.dense(x,1024)
h1_1=tf.nn.relu(h1)
h2 = tf.layers.dense(h1_1,1024)
h2_1=tf.nn.relu(h2)
y = tf.layers.dense(h2_1,513)
y_1 = tf.nn.relu(y)
```

Define a place holder for the output, mean squared error to calculate loss and gradient descent optimizer with learning rate 0.1

```python
y_ = tf.placeholder(tf.float32, [None,513])
mean_squared = tf.losses.mean_squared_error(y_,y_1)
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(mean_squared)
loss_log = []
init = tf.global_variables_initializer()
epochs=500
```

Split data into 100 batches of 25 each.

```python
batch_S=np.array_split(S_abs,100)
batch_X=np.array_split(X_abs,100)
batches=100
```

Train the model

```python
new_sess=tf.Session()
new_sess.run(init) # reset values to wrong
    # training loop    



for epoch in range(epochs):
    print("epoch",epoch)
    temp_loss=0
    for _ in range(batches):
        batch_xs, batch_ys = batch_X[_],batch_S[_]
        _,loss = new_sess.run([train_step,mean_squared], feed_dict={x: batch_xs, y_: batch_ys})
        
        temp_loss+=loss
    loss_log.append(temp_loss/batches)
    print("loss",temp_loss/batches)
```

Test the model on test audio 1.
Load the test file, apply stft and take magnitude of it. 
Run the model on the test file, convert the output back to normal using istft.

```python
test1, sr=librosa.load("test_x_01.wav", sr=None)
test1=librosa.stft(test1, n_fft=1024, hop_length=512)
test1_abs=np.abs(np.transpose(test1))
```

```python
test1_output= new_sess.run(y_1, feed_dict={x:test1_abs})
test1_output.shape
```

```python
clean_test1=(test1/np.abs(test1))*(np.transpose(test1_output))
clean_test1.shape
clean_test1_istft=librosa.istft(clean_test1, hop_length=512)
clean_test1_istft.shape
librosa.output.write_wav("test_s_01_recons.wav", clean_test1_istft, sr)
```

Test file 2.

```python
test2, sr=librosa.load("test_x_02.wav", sr=None)
test2=librosa.stft(test2, n_fft=1024, hop_length=512)
test2_abs=np.abs(np.transpose(test2))
test2_output= new_sess.run(y_1, feed_dict={x:test2_abs})
clean_test2=(test2/np.abs(test2))*(np.transpose(test2_output))
clean_test2.shape
clean_test2_istft=librosa.istft(clean_test2, hop_length=512)
clean_test2_istft.shape
librosa.output.write_wav("test_s_02_recons.wav", clean_test2_istft, sr)
```
